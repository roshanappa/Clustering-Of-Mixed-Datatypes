- If you have ecommerce kind of business and you want to send the promotional offer to the some standard customer not to all customer.Rather you want to cluster the customer in different bucket based on the demographic and based on the product behaviour and you wanted to targeted offer to each cluster of customers.In this case you may get the combination of categorical data and continuous data.

- In the case of clustering,we always prefer the K-means clustering,it help us to get the cluster of the continuous data by using the eucleadian distance and data points mean.

- In this scinario the categorical data is not applicable and if we use by using labelencoding or onehotending even it does not makes any sense.

- For that we have another algorithm this called as `KPrototypes`

```python
import numpy as np
import pandas as pd
```

```python
from kmodes.kprototypes import KPrototypes
```

```python
marketing_df = pd.read_csv('https://raw.githubusercontent.com/srivatsan88/YouTubeLI/master/dataset/marketing_cva_f.csv')
```

```python
marketing_df
```

There are so many columns are avaliable and each column having the different datatypes.So in order reduce the complexity we are going foreward and selecting few columns for simplicity.

```python
marketing_df = marketing_df.drop(columns=['Customer','Vehicle_Class','avg_vehicle_age','months_last_claim','Total_Claim_Amount'])
```

```python
marketing_df
```

Now we are going to convert pandas dataframe into the numpy array.

```python
mark_array = marketing_df.to_numpy()
```

Now we are going to convert the each numerical column in float datatypes.

```python
mark_array[:,1] = mark_array[:,1].astype('float')
mark_array[:,3] = mark_array[:,3].astype('float')
mark_array[:,5] = mark_array[:,5].astype('float')
mark_array[:,6] = mark_array[:,6].astype('float')
```

```python
mark_array
```

Now,we can see that all the numrical datatype is getting converted into the float data type format.

In the below we have use the algorithm and in that we have first go for the 3 clsuter and uses the number of iteration 100 and while fit_predict method we have decleare over there that we having the categorical variable at the 0th,2nd and 4th index position of columns.

```python
kproto = KPrototypes(n_clusters=3,verbose=2,n_jobs=-1,max_iter=100,random_state=42)
cluster = kproto.fit_predict(mark_array,categorical=[0,2,4])
```

```python
print(kproto.cluster_centroids_)
```

```python
cluster_lst =[]

for i in cluster:
  cluster_lst.append(i)
```

```python
cluster_lst
```

Now we are going to asssign the cluster to the original dataframe.

```python
marketing_df['Cluster'] = cluster_lst
```

Now one by one we are going to look for each clsuter data.

```python
marketing_df[marketing_df['Cluster'] ==0].head(10)
```

```python
marketing_df[marketing_df['Cluster'] ==1].head(10)
```

```python
marketing_df[marketing_df['Cluster'] ==2].head(10)
```
